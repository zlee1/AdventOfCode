## Part 1

Something is wrong with global snow production, and you've been selected to take a look. The Elves have even given you a map; on it, they've used stars to mark the top fifty locations that are likely to be having problems.

You've been doing this long enough to know that to restore snow operations, you need to check all fifty stars by December 25th.

Collect stars by solving puzzles. Two puzzles will be made available on each day in the Advent calendar; the second puzzle is unlocked when you complete the first. Each puzzle grants one star. Good luck!

You try to ask why they can't just use a weather machine ("not powerful enough") and where they're even sending you ("the sky") and why your map looks mostly blank ("you sure ask a lot of questions") and hang on did you just say the sky ("of course, where do you think snow comes from") when you realize that the Elves are already loading you into a trebuchet ("please hold still, we need to strap you in").

As they're making the final adjustments, they discover that their calibration document (your puzzle input) has been amended by a very young Elf who was apparently just excited to show off her art skills. Consequently, the Elves are having trouble reading the values on the document.

The newly-improved calibration document consists of lines of text; each line originally contained a specific calibration value that the Elves now need to recover. On each line, the calibration value can be found by combining the first digit and the last digit (in that order) to form a single two-digit number.

For example:

>1abc2
>
>pqr3stu8vwx
>
>a1b2c3d4e5f
>
>treb7uchet


In this example, the calibration values of these four lines are 12, 38, 15, and 77. Adding these together produces 142.

Consider your entire calibration document. What is the sum of all of the calibration values?

```python
with open('data/input.txt') as f:
    data = f.readlines()
```

```python
def get_first(line):
    # loop through line and return first numeric character
    for i in line:
        if(i.isnumeric()):
            return int(i)
```

```python
def get_last(line):
    # loop through line backwards and return first (technically last) numeric character
    for i in line[::-1]:
        if(i.isnumeric()):
            return int(i)
```

```python
# create list of all calibration numbers
calibration_nums = []
for i in data:
    # calibration numbers are a combination of the first and last digits of the line
    calibration_nums.append(int(str(get_first(i)) + str(get_last(i))))
```

```python
# output sum of calibration numbers
print(sum(calibration_nums))
```

# Part 2

Your calculation isn't quite right. It looks like some of the digits are actually spelled out with letters: one, two, three, four, five, six, seven, eight, and nine also count as valid "digits".

Equipped with this new information, you now need to find the real first and last digit on each line. For example:

>two1nine
>
>eightwothree
>
>abcone2threexyz
>
>xtwone3four
>
>4nineeightseven2
>
>zoneight234
>
>7pqrstsixteen

In this example, the calibration values are 29, 83, 13, 24, 42, 14, and 76. Adding these together produces 281.

What is the sum of all of the calibration values?

```python
def replace_str_nums(line):
    # replace number words with their actual values
    
    indices = {} # dict of starting indices of each word-number
    
    # conversion dict for word-number to value
    convert = {'one':'1', 
               'two':'2',
               'three':'3',
               'four':'4',
               'five':'5',
               'six':'6',
               'seven':'7',
               'eight':'8',
               'nine':'9'
              }
    
    # loop through all word-numbers
    for key in convert.keys():
        # add a key:value pair to indices
        # key = starting index of word-number
        # value = word-number
        ids = [index for index in range(len(line)) if line.startswith(key, index)]
        for ind in ids:
            indices.update({ind:key})
    
    # sort list of indices that mark the beginning of a word-number
    indices_list = list(indices.keys())
    indices_list.sort()
        
    new_line = '' # empty string to be appended to with converted numbers
    
    # marks how many characters remain in a word-number within loop
    reached = 0 
    # ex: line=omtwosp; reached values shown are at end of iteration
    # iteration 1: char=o, reached=0, append=o
    # iteration 2: char=m, reached=0, append=m
    # iteration 3: char=t, reached=2, append=2 (numerical value of word-number)
    # iteration 4: char=w, reached=1, append=''
    # iteration 5: char=o, reached=0, append=''
    # characters can only be appended when reached == 0 at start of iteration
    # this allows for correct updating of lines with overlapping word-numbers
    # ex: twone -> 21 instead of 2ne
    for i, ch in enumerate(line):
        # marks that word-number was found
        # append numerical value and update reached with word length
        if(i in indices.keys()):
            new_line += convert.get(indices.get(i))
            reached = len(indices.get(i))-1
        # decrement reached
        elif(reached > 0):
            reached -= 1
        # append character of current iteration when reached == 0
        else:
            new_line += ch
            
    # return newly created line with word-number conversions applied
    return new_line
```

```python
# create list of all calibration numbers
calibration_nums = []
for i in data:
    replaced = replace_str_nums(i)
        # calibration numbers are a combination of the first and last digits of the line
    calibration_nums.append(int(str(get_first(replaced)) + str(get_last(replaced))))
print(sum(calibration_nums))
```
