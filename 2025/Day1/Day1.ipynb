# Advent of Code 2025 Day 1: Secret Entrance

## Part 1

```python
import math
```

```python
values = []
with open('input.txt', 'r') as f:
    for line in f.readlines():
        values.append((1 if line.replace('\n', '')[0] == 'R' else -1)*int(line.replace('\n', '')[1:]))
```

```python
print(values)
```

```python
total = 50
zeroes = 0
for i in range(len(values)):
    total += values[i]
    if(total%100 == 0):
        zeroes += 1

print(zeroes)
```

## Part 2

```python
total = 50
zeroes = 0
for i in range(len(values)):
    for j in range(abs(values[i])):
        total += 1 if values[i] >= 0 else -1
        if(total%100 == 0):
            zeroes += 1
            total = 0

print(zeroes)
```
