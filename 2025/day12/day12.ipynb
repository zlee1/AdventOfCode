```python
with open('input.txt', 'r') as f:
    data = f.read().split('\n')[:-1]
```

```python
presents = {}
for i in range(6):
    present_idx = data.index(f'{i}:')
    presents[i] = list(map(list, data[present_idx+1:present_idx+4]))

# print(presents)

regions = []
for i in range(data.index('5:')+5, len(data)):
    dimensions = list(map(int, data[i].split(': ')[0].split('x')))
    present_quantities = {}
    for index, quantity in enumerate(list(map(int, data[i].split(': ')[1].split(' ')))):
        present_quantities[index] = quantity
    
    regions.append([dimensions, present_quantities])

# print(regions)
```

```python
# keeping this stuff because it might be in part 2 idk
import copy
import numpy as np

def create_grid(dimensions):
    return [['.' for i in range(dimensions[0])] for j in range(dimensions[1])]

def print_grid(grid):
    for row in grid:
        print(' '.join(row))

def place_present(grid, row, col, present, rotations=0, character='#'):
    rotated_present = np.rot90(present, rotations*-1)

    new_grid = copy.deepcopy(grid)
    for p_row in range(len(rotated_present)):
        for p_col in range(len(rotated_present[p_row])):
            if(grid[row+p_row][col+p_col] == '.'):
                if(rotated_present[p_row][p_col] == '#'):
                    new_grid[row+p_row][col+p_col] = character
            else:
                print(f'present overlaps at ({row}, {col})')
                return grid
    return new_grid

def get_present_points(present, rotations=0):
    rotated_present = np.rot90(present, rotations*-1)

    points = set()
    for row in range(len(rotated_present)):
        for col in range(len(rotated_present[0])):
            if(rotated_present[row][col] == '#'):
                points.add((row, col))
    return points

# def check_fits(grid, row, col, present):
#     valid_rotations = []
#     for i in range(3):

#     return

grid = create_grid(regions[0][0])
# print_grid(place_present(grid, 0, 0, presents[4], rotations=1))

# print(get_present_points(presents[4], rotations=1))
```

```python
# ugh.

present_sizes = {}
for present in presents:
    present_size = 0
    for row in presents[present]:
        for col in row:
            if(col == '#'):
                present_size += 1
    present_sizes[present] = present_size

valid_area = 0
for region in regions:
    area = region[0][0]*region[0][1]
    present_area = 0
    for i in region[1]:
        present_area += present_sizes[i]*region[1][i]
    if(area > present_area):
        valid_area += 1

# print(valid_area)
```
