## Part 1

```python
import re
import numpy as np
```

```python
with open('input.txt', 'r') as f:
    # splitting without leaving empty strings behind is surprisingly annoying
    data = [re.sub(r'[\s]{2,}', ' ', row.strip()).split(' ') for row in f.read().split('\n')[:-1]]
```

```python
# flip it and math it
data_t = np.transpose(data)
total = 0
for i in data_t:
    total += eval(i[-1].join(i[:-1]))

# print(total)
```

## Part 2

```python
# oops, i misjudged the problem in part 1
with open('input.txt', 'r') as f:
    data = f.read().split('\n')[:-1]
```

```python
total = 0

nums = []
cur_op = ''
for i in range(len(data[0])):
    # append the concatenation of a column, excluding operators
    cur_num = ''.join([row[i] if row[i] not in ('*', '+') else ' ' for row in data]).strip()
    if(cur_num != ''):
        nums.append(cur_num)
        cur_op = data[-1][i] if data[-1][i] in ('*', '+') else cur_op

    # calculate result when blank column is reached and reset vars
    if(cur_num == '' or i == len(data[0])-1):
        total += eval(cur_op.join(nums))
        nums = []
        cur_op = ''
    
# print(total)
```
