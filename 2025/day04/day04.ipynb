## Part 1

```python
with open('input.txt', 'r') as f:
    data = f.read().split('\n')[:-1]
```

```python
data = [list(row) for row in data]
```

```python
def check_surroundings(data, row, col, symbol, limit):
    symbol_ct = 0
    for i in range(9):
        new_row = row+((i%3)-1)
        new_col = col+((i//3)-1)
        if(new_row >= 0 and new_col >= 0 and new_row < len(data) and new_col < len(data[row]) and not ((i%3)-1 == 0 and (i//3)-1 == 0)):
            if(data[new_row][new_col] == symbol):
                symbol_ct += 1
        
        if(symbol_ct >= limit):
            return 0
    return 1


total = 0
for row in range(len(data)):
    for col in range(len(data[row])):
        if(data[row][col] == '@'):
            total += check_surroundings(data, row, col, '@', 4)

# print(total)
```

## Part 2

```python
def check_surroundings(data, row, col, symbol, limit):
    symbol_ct = 0
    for i in range(9):
        new_row = row+((i%3)-1)
        new_col = col+((i//3)-1)

        if(new_row >= 0 and new_col >= 0 and new_row < len(data) and new_col < len(data[row]) and not ((i%3)-1 == 0 and (i//3)-1 == 0)):
            if(data[new_row][new_col] == symbol):
                symbol_ct += 1
        
        if(symbol_ct >= limit):
            return 0
    return 1


removed = 1
while removed:
    removed = 0
    
    for row in range(len(data)):
        for col in range(len(data[row])):
            if(data[row][col] == '@'):
                if(check_surroundings(data, row, col, '@', 4)):
                    removed += 1
                    data[row][col] = 'X'

# print(sum([row.count('X') for row in data]))
```
