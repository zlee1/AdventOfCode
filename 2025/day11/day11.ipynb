```python
import copy

# get all devices that are directly connected to end device
def get_devices_to(end, devices):
    valid_devices = []
    for device, outputs in devices.items():
        if(end in outputs):
            valid_devices.append(device)
            continue
    return valid_devices

# get all devices that can lead to end device
def get_valid_before(end, devices):

    valid_before = {end}
    checked_devices = set()
    change = 1

    while len(checked_devices) < len(valid_before):
        cur_device = list(valid_before-checked_devices)[0]

        for device in get_devices_to(cur_device, devices):
            if(device not in valid_before):
                valid_before.add(device)

        checked_devices.add(cur_device)
    return valid_before

def find_path(start, end, devices, required_points=[], excluded_points=[]):
    
    global before_fft
    global before_dac
    global before_out
    global points

    # all points that lead to each of the required points in the path
    if(before_fft == set()):
        before_fft = get_valid_before('fft', devices)
    if(before_dac == set()):
        before_dac = get_valid_before('dac', devices)
    if(before_out == set()):
        before_out = get_valid_before('out', devices)
    
    num_paths = 0
    if(start == end):
        for i in required_points:
            if(start != i and i in devices):
                return 0
            
        return 1
    if(devices == {} or devices is None or start is None or start not in devices):
        return 0
    else:
        for i in devices.get(start):

            if(i in excluded_points):
                continue

            if(i in points):
                num_paths += points.get(i)
                continue
            
            if('fft' in devices and i not in before_fft and start != 'fft' and 'fft' in required_points):
                continue
            if('dac' in devices and i not in before_dac and start != 'dac' and 'dac' in required_points):
                continue
            if('out' in devices and i not in before_out and start != 'out' and 'out' in required_points):
                continue

            new_devices = copy.deepcopy(devices)
            new_devices.pop(start)

            segment_paths = find_path(i, end, new_devices, required_points, excluded_points)
            num_paths += segment_paths
            if(i not in points):
                points[i] = segment_paths

    return num_paths

devices = {}
for device in data:
    spl = device.split(' ')
    devices[spl[0][:-1]] = spl[1:]

# i did this whole big brain segmenting thing where i find the number of paths for each individual segment of the whole but then when i implemented memoization it didn't even matter because memoization method is so fast
# points = {}
# svr_fft = find_path_segment('svr', 'fft', devices, required_points=['fft'], excluded_points=['dac'])
# points = {}
# fft_dac = find_path_segment('fft', 'dac', devices, required_points=['fft', 'dac'], excluded_points=[])
# points = {}
# dac_out = find_path_segment('dac', 'out', devices, required_points=['dac', 'out'], excluded_points=['fft'])

# points = {}
# svr_dac = find_path_segment('svr', 'dac', devices, required_points=['dac'], excluded_points=['fft'])
# points = {}
# dac_fft = find_path_segment('dac', 'fft', devices, required_points=['fft', 'dac'], excluded_points=[])
# points = {}
# fft_out = find_path_segment('fft', 'out', devices, required_points=['fft', 'out'], excluded_points=['dac'])
# print((svr_fft*fft_dac*dac_out)+(svr_dac*dac_fft*fft_out))



points = {}
total = find_path('svr', 'out', devices, required_points=['fft', 'dac'])
# print(total)
```
