## Part 1

```python
with open('input.txt', 'r') as f:
    data = f.read().split('\n')[:-1]
```

```python
total = 0
for bank in data:
    max_val = max(list(bank))
    if(bank.index(max_val) == len(bank)-1):
        low_val = max_val
        new_max = max(list(bank.replace(max_val, '')))
        max_val = new_max
    else:
        low_val = max(list(bank)[bank.index(max_val)+1:])

    total += int(''.join([max_val, low_val]))
    
#print(total)
```

```python
# updated part 1 to use part 2 solution
joltage_len = 2
total = 0
for bank in data:
    joltage = []
    cur_index = 0

    while len(joltage) < joltage_len:
        remaining_nums = joltage_len-len(joltage)
        last_possible_index = len(bank)-remaining_nums+1
        # get max from bank within range of most recent max and last possible index that is at least the remaining length long
        cur_max = max(bank[cur_index:last_possible_index])

        joltage.append(cur_max)

        cur_index = bank[cur_index:last_possible_index].index(cur_max) + cur_index + 1
    total += int(''.join(joltage))

# print(total)
```
