# Day 02: Gift Shop

## Part 1

```python
import re
```

```python
with open('input.txt', 'r') as f:
    data = f.read().split(',')
```

```python
total = 0
for range_set in data:
    low_val = int(range_set.split('-')[0])
    high_val = int(range_set.split('-')[1])

    for i in range(low_val, high_val+1):
        num_len = len(str(i))
        if(len(str(i))%2 == 0):
            if(str(i)[:num_len//2] == str(i)[num_len//2:]):
                total += i

# print(total)
```

## Part 2

```python
total = 0
for range_set in data:
    low_val = int(range_set.split('-')[0])
    high_val = int(range_set.split('-')[1])

    for i in range(low_val, high_val+1):
        for j in range(1, (len(str(i))//2)+1):
            if(len(str(i))%j == 0):
                if(re.match('(' + str(i)[:j] + '){2,}\\b', str(i))):
                    total += i
                    break

#print(total)
```
