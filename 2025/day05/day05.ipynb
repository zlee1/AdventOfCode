## Part 1

```python
with open('input.txt', 'r') as f:
    data = f.read().split('\n')[:-1]
```

```python
ranges = data[:data.index('')]
ids = [int(id) for id in data[data.index('')+1:]]

fresh = []
for id in ids:
    for id_range in ranges:
        range_start = int(id_range.split('-')[0])
        range_end = int(id_range.split('-')[1])
        if(id >= range_start and id <= range_end and id not in fresh):
            fresh.append(id)
            break
        
# print(len(fresh))
```

## Part 2

```python
def consolidate(range_dict: dict, changes: int):
    if(changes == 0):
        return range_dict
    
    new_dict = range_dict.copy()

    changes = 0
    for range_start, range_end in range_dict.items():
        cur_changes = changes
        for comp_range_start, comp_range_end in range_dict.items():
            if((range_start == comp_range_start and range_end == comp_range_end) or comp_range_start not in new_dict.keys()):
                continue
            elif(range_start not in new_dict.keys()):
                break
            elif(range_end < comp_range_start or range_start > comp_range_end):
                continue
            elif(range_start >= comp_range_start and range_end <= comp_range_end):
                # print(f'{range_start}-{range_end} within {comp_range_start}-{comp_range_end}, cur removed')

                new_dict.pop(range_start)
                new_dict.pop(comp_range_start)

                new_dict[comp_range_start] = comp_range_end
                
                changes += 1
                break
            # lower start with end greater than or equal to original start
            elif(range_start <= comp_range_start and range_end >= comp_range_start):
                # if end is greater than original, new start and end
                if(range_end >= comp_range_end):
                    # print(f'{range_start}-{range_end} encompasses {comp_range_start}-{comp_range_end}, comp removed')
                    
                    new_dict.pop(range_start)
                    new_dict.pop(comp_range_start)
                    
                    new_dict[range_start] = range_end

                    changes += 1
                    break
                else:
                    # if end is not greater than original, new start
                    # print(f'{range_start}-{range_end} earlier start {comp_range_start}-{comp_range_end}, comp removed, cur updated to {range_start}-{comp_range_end}')

                    new_dict.pop(range_start)
                    new_dict.pop(comp_range_start)

                    new_dict[range_start] = comp_range_end

                    changes += 1
                    break
            # higher end with start less than or equal to original end
            elif(range_end >= comp_range_end and range_start <= comp_range_end):
                # if start less than original start, new start and end
                if(range_start <= comp_range_start):
                    # print(f'{range_start}-{range_end} encompasses {comp_range_start}-{comp_range_end}, comp removed')

                    new_dict.pop(range_start)
                    new_dict.pop(comp_range_start)

                    new_dict[range_start] = range_end

                    changes += 1
                    break
                # if start not less than original, new end
                else:
                    # print(f'{range_start}-{range_end} later end {comp_range_start}-{comp_range_end}, cur removed, comp updated to {comp_range_start}-{range_end}')
                    
                    new_dict.pop(range_start)
                    new_dict.pop(comp_range_start)

                    new_dict[comp_range_start] = range_end
                    
                    changes += 1
                    break

        if(cur_changes != changes):
            break
    
    return consolidate(new_dict, changes)

ranges = data[:data.index('')]

range_dict = {}
for id_range in ranges:
    range_start = int(id_range.split('-')[0])
    range_end = int(id_range.split('-')[1])
    if(range_start in range_dict.keys()):
        if(range_end >= range_dict[range_start]):
            range_dict.pop(range_start)
            range_dict[range_start] = range_end
        else:
            continue
    else:
        range_dict[range_start] = range_end

consolidated = consolidate(range_dict, 1)

# print(sum([(range_end-range_start)+1 for range_start, range_end in consolidated.items()]))
```
