## Part 1

```python
with open('input.txt', 'r') as f:
    data = f.read().split('\n')[:-1]
```

```python
data = [
'3-5'
,'10-14'
,'16-20'
,'12-18'
,''
,'1'
,'5'
,'8'
,'11'
,'17'
,'32'
]
```

```python
ranges = data[:data.index('')]
ids = [int(id) for id in data[data.index('')+1:]]

fresh = []
for id in ids:
    for id_range in ranges:
        range_start = int(id_range.split('-')[0])
        range_end = int(id_range.split('-')[1])
        if(id >= range_start and id <= range_end and id not in fresh):
            fresh.append(id)
            break
print(len(fresh))
```

```python
def consolidate(range_dict: dict, changes: int):
    if(changes == 0):
        return range_dict
    

    new_dict = range_dict.copy()
    changes = 0
    for range_start, range_end in range_dict.items():
        for comp_range_start, comp_range_end in range_dict.items():
            if(range_start >= comp_range_start and range_end <= comp_range_end):
                new_dict.pop(range_start)
                changes += 1
                break
            # lower start with end greater than or equal to original start
            elif(range_start <= comp_range_start and range_end >= comp_range_start):
                # if end is greater than original, new start and end
                if(range_end >= comp_range_end):
                    range_dict.pop(comp_range_start)
                    changes += 1
                    break
                else:
                    # if end is not greater than original, new start
                    range_dict.pop(comp_range_start)
                    range_dict[range_start] = comp_range_end
                    changes += 1
                    break
            # higher end with start less than or equal to original end
            elif(range_end >= comp_range_end and range_start <= comp_range_end):
                # if start less than original start, new start and end
                if(range_start <= comp_range_start):
                    range_dict.pop(comp_range_start)
                    changes += 1
                    break
                # if start not less than original, new end
                else:
                    range_dict[comp_range_start] = range_end
                    changes += 1
                    break
    
    return consolidate(new_dict, changes)

ranges = data[:data.index('')]

range_dict = {int(id_range.split('-')[0]): int(id_range.split('-')[1]) for id_range in ranges}

print(consolidate(range_dict, 1))
```
