```python
def find_options(data: list, row: int, col: int):
    val = data[row][col]

    options = []
    for ch_row, ch_col in [[-1, 0], [1, 0], [0, -1], [0, 1]]:
        cur_row = row+ch_row
        cur_col = col+ch_col
        if(cur_row >= 0 and cur_row < len(data) and cur_col >= 0 and cur_col < len(data[0])):
            if(data[cur_row][cur_col] == val):
                options.append((cur_row, cur_col))
    # print(f'options for ({row}, {col}):', options)
    return options
```

```python
def find_region(data, row, col):
    val = data[row][col]

    options = find_options(data, row, col)
    visited = [(row, col)]
    while options != []:
        cur_row, cur_col = options.pop(0)

        if((cur_row, cur_col) not in visited):
            options += find_options(data, cur_row, cur_col)
            visited.append((cur_row, cur_col))
    return visited
```

```python
def get_perimeter(data: list, region: list):
    perimeter = 0
    for row, col in region:
        for ch_row, ch_col in [[-1, 0], [1, 0], [0, -1], [0, 1]]:
            cur_row = row+ch_row
            cur_col = col+ch_col

            if(cur_row < 0 or cur_row >= len(data) or cur_col < 0 or cur_col >= len(data[0]) or data[cur_row][cur_col] != data[row][col]):
                perimeter += 1
    return perimeter
            
```

```python
def get_adjacent(data: list, row: int, col: int):
    directions = ['top', 'bottom', 'left', 'right']
    changes = [[-1, 0], [1, 0], [0, -1], [0, 1]]

    adjacents = {}
    for i in range(4):
        ch_row, ch_col = changes[i]
        cur_row = row+ch_row
        cur_col = col+ch_col

        if(cur_row >= 0 and cur_row < len(data) and cur_col >= 0 and cur_col < len(data[0])):
            adjacents[directions[i]] = data[cur_row][cur_col]
        else:
            adjacents[directions[i]] = None
    
    return adjacents
```

```python
def get_sides(data: list, region: list):
    sides = []
    
    unplaced = region[:]
    while unplaced:
        row, col = unplaced.pop(0)
        adjacents = get_adjacent(data, row, col)

        for direction, value in adjacents.items():
            
            if(value != data[row][col]):
                sides.append([[(row, col)], direction])
    
    return sides
```

```python
def consolidate_sides(sides: list):
    consolidated_sides = []

    for coords, direction in sides:
        for row, col in coords:
            # for every consolidated side
            appended = False
            for i in consolidated_sides:
                # print(i)
                if(i[1] == direction):
                    # break if pair already exists in consolidated side
                    if((row, col) in i[0]):
                        break
                    else:
                        for row2, col2 in i[0]:
                            if(direction in ['top', 'bottom'] and abs(col2-col) == 1 and row2 == row):
                                i[0].append((row, col))
                                appended = True
                                # print('appending', row, col, direction, 'to', i)
                                break
                            elif(direction in ['left', 'right'] and abs(row2-row) == 1 and col2 == col):
                                i[0].append((row, col))
                                appended = True
                                # print('appending', row, col, direction, 'to', i)
                                break
            if(not appended):
                consolidated_sides.append([[(row, col)], direction])
    return consolidated_sides
```
