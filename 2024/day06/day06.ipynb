## Part 1

```python
def get_start(data):
    for row in range(len(data)):
        for col in range(len(data[row])):
            if(data[row][col] == '^'):
                return row, col
```

```python
def check_valid_spot(data, row, col):
    return data[row][col] == '.' or data[row][col] == 'X'
```

```python
def rotate(data, row, col):
    if(data[row][col] == '^'):
        data[row][col] = '>'
    elif(data[row][col] == '>'):
        data[row][col] = 'v'
    elif(data[row][col] == 'v'):
        data[row][col] = '<'
    elif(data[row][col] == '<'):
        data[row][col] = '^'
    return data
```

```python
def get_next_pos(data, row, col):
    if(data[row][col] == '^'):
        return row-1, col
    elif(data[row][col] == '>'):
        return row, col+1
    elif(data[row][col] == 'v'):
        return row+1, col
    elif(data[row][col] == '<'):
        return row, col-1
    
```

```python
def move(data, cur_row, cur_col):
    try:
        rotated = 0
        new_row, new_col = get_next_pos(data, cur_row, cur_col)

        if(new_row < 0 or new_row > len(data)-1 or new_col < 0 or new_col > len(data[0])-1):
            data[cur_row][cur_col] = 'X'
            return data, cur_row, cur_col, 1, rotated

        # print(data[cur_row][cur_col], data[new_row][new_col])
        if(check_valid_spot(data, new_row, new_col)):
            # print('valid - swapping')
            data[new_row][new_col] = data[cur_row][cur_col]
            data[cur_row][cur_col] = 'X'
            
            return data, new_row, new_col, 0, rotated
        else:
            # print('invalid - rotating')
            data = rotate(data, cur_row, cur_col)
            rotated = 1
            # for i in data:
                # print(i)
            # print('data')
            return data, cur_row, cur_col, 0, rotated
    except IndexError:
        data[cur_row][cur_col] = 'X'
        return data, cur_row, cur_col, 1, rotated
```
