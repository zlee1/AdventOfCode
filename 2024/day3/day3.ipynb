```python
with open('data\\input.txt', 'r') as f:
    data = f.readlines()
```

```python
data = ''.join(data)
```

```python
import re
muls = re.findall('mul\([0-9]{1,3},[0-9]{1,3}\)', data)
```

```python
total = 0
for i in muls:
    nums = list(map(int, i.replace('mul(', '').replace(')', '').split(',')))

    total += nums[0]*nums[1]

total
```

```python
mul_exp = re.compile("mul\([0-9]{1,3},[0-9]{1,3}\)")

indices = []
groups = []
for m in mul_exp.finditer(data):
    indices.append(m.start())
    groups.append(m.group())

do_exp = re.compile("do\(\)")

do_indices = []
do_groups = []
for do in do_exp.finditer(data):
    do_indices.append(do.start())
    do_groups.append(do.group())

dont_exp = re.compile("don't\(\)")

dont_indices = []
dont_groups = []
for dont in dont_exp.finditer(data):
    dont_indices.append(dont.start())
    dont_groups.append(dont.group())


total = 0
do = 1
for i in range(max(indices)+1):
    if(i in do_indices):
        do = 1
    
    if(i in dont_indices):
        do = 0
    
    if(do == 1 and i in indices):
        nums = list(map(int, groups[indices.index(i)].replace('mul(', '').replace(')', '').split(',')))

        total += nums[0]*nums[1]

total
```
