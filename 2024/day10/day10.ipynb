```python
def find_moves(data: list, row: int, col: int):
    valid_moves = []
    for ch_row, ch_col in [[-1, 0], [1, 0], [0, -1], [0, 1]]:
        cur_row = row+ch_row
        cur_col = col+ch_col
        if(cur_row >= 0 and cur_row < len(data) and cur_col >= 0 and cur_col < len(data[0])):
            if(data[cur_row][cur_col] == data[row][col]+1):
                valid_moves.append((cur_row, cur_col))
    # print(f'valid moves for ({row}, {col}):', valid_moves)
    return valid_moves
```

```python
def find_paths(data: list, row: int, col: int):
    found_paths = []
    moves = find_moves(data, row, col)
    visited = []
    while moves != []:
        cur = moves.pop(0)
        cur_row = cur[0]
        cur_col = cur[1]
        if((cur_row, cur_col) not in visited):
            #visited.append((cur_row, cur_col)) # commented out to solve for part 2
            if(data[cur_row][cur_col] == 9):
                found_paths.append((cur_row, cur_col))
        else:
            pass
            # print(f'Already visited ({cur_row}, {cur_col})')

        moves += find_moves(data, cur_row, cur_col)

    return found_paths
```

```python
with open('data\\input.txt', 'r') as f:
    data = [list(map(int, i)) for i in f.read().splitlines()]

trailheads = []
for row in range(len(data)):
    for col in range(len(data[0])):
        if(data[row][col] == 0):
            trailheads.append(find_paths(data, row, col))

print(sum([len(i) for i in trailheads]))
```
