## Part 1

```python
def apply_rules(stone: int):
    str_stone = str(stone)
    if(stone == 0):
        return [1]
    elif(len(str_stone)%2 == 0):
        return [int(str_stone[:len(str_stone)//2]), int(str_stone[len(str_stone)//2:])]
    else:
        return [stone*2024]
```

```python
def blink(data: list):
    new_data = []
    for stone in data:
        new_data += apply_rules(stone)
    
    return new_data
```

```python
with open('data\\input.txt', 'r') as f:
    data = [int(i) for i in f.read().split(' ')]

print(data)

for i in range(1, 26):
    data = blink(data)
    print(i, len(data))#sorted(data)[:100])
```

## Part 2

```python
def blink_dict(data: dict):
    new_data = {}
    for stone, count in data.items():
        new_stones = apply_rules(stone)
        for new_stone in new_stones:
            if(new_stone in new_data.keys()):
                new_data[new_stone] += count
            else:
                new_data[new_stone] = count
    return new_data
```

```python
with open('data\\input.txt', 'r') as f:
    raw_data = [int(i) for i in f.read().replace('\n', '').split(' ')]

data = {}
for i in raw_data:
    data[i] = raw_data.count(i)

for i in range(1, 76):
    data = blink_dict(data)
    print(i, sum(data.values()))
```
