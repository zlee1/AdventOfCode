# Day 5

```python
with open('data\\input.txt', 'r') as f:
    data = f.read().splitlines()
```

## Part 1

```python
import math
import random
```

```python
rules = [list(map(int, i.split('|'))) for i in data[:data.index('')]]
updates = [list(map(int, i.split(','))) for i in data[data.index('')+1:]]
```

```python
def check_broken(update, rules):
    for rule in rules:
        try:
            if(update.index(rule[0]) > update.index(rule[1])):
                return 1
        except ValueError as e:
            pass
    return 0
```

```python
total = 0
for update in updates:
    mid_val = update[math.floor(len(update)/2)]
    if(not check_broken(update, rules)):
        total += mid_val

total
```

## Part 2

```python
def apply_rules(update, rules):
    while check_broken(update, rules):
        random.shuffle(rules)
        for rule in rules:
            if(update.index(rule[0]) > update.index(rule[1])):
                update[update.index(rule[0])] = rule[1]
                update[update.index(rule[1])] = rule[0]
    return update
```

```python
total = 0
for update in updates:
    if(check_broken(update, rules)):
        applicable_rules = [i for i in rules if i[0] in update and i[1] in update]

        ordered_update = apply_rules(update.copy(), applicable_rules)

        total += ordered_update[math.floor(len(ordered_update)/2)]
total
```
