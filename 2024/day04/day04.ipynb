```python
with open('data\\input.txt', 'r') as f:
    data = f.readlines()
```

```python
def check_xmas(data, row, col, cur_char, row_change=None, col_change=None):
    next_char = 'XMAS'['XMAS'.index(cur_char)+1]

    total = 0
    if(row_change is None and col_change is None):
        for i in range(9):
            row_change = int(i/3)-1
            col_change = i%3-1

            if(row+row_change < 0 or row+row_change >= len(data) or col+col_change < 0 or col+col_change >= len(data[0])):
                continue
            elif(data[row+row_change][col+col_change] == next_char):
                total += check_xmas(data, row+row_change, col+col_change, next_char, row_change=row_change, col_change=col_change)
    elif(row+row_change >= 0 and row+row_change < len(data) and col+col_change >= 0 and col+col_change < len(data[0]) and data[row+row_change][col+col_change] == next_char):
        if(next_char == 'S'):
            total += 1
        else:
            total += check_xmas(data, row+row_change, col+col_change, next_char, row_change=row_change, col_change=col_change)
    return total

```

```python
total = 0
for row in range(len(data)):
    for col in range(len(data[row])):
        if(data[row][col] == 'X'):
            total += check_xmas(data, row, col, 'X')
total
```
