## Part 1 & 2

```python
def evaluate(num, values, operators, options=['*', '+', '||']):

    cur_val = values[0]

    for i in range(len(operators)):
        if(options[operators[i]] == '||'):
            cur_val = int(str(cur_val)+str(values[i+1]))
        else:
            cur_val = eval(f'{cur_val}{options[operators[i]]}{values[i+1]}')

    if(cur_val == num):
        return 1
    else:
        return 0
```

```python
def find_solution(num, values, options=['*', '+', '||']):
    
    for i in itertools.product(range(len(options)), repeat=len(values)-1):
        if(evaluate(num, values[:], list(i)[:]) == 1):
            return 1

    return 0
```

```python
total = 0
for i in data:
    num = int(i.split(': ')[0])
    values = list(map(int, i.split(': ')[1].split(' ')))
    
    if(find_solution(num, values) == 1):
        total += num
total

```
