```python
def is_valid(data, point):
    return point[0] >= 0 and point[0] <= len(data)-1 and point[1] >= 0 and point[1] <= len(data[0])-1
```

```python
def get_antinode_points(data, antennae):
    vdiff = antennae[0][0]-antennae[1][0]
    hdiff = antennae[0][1]-antennae[1][1]

    points = []

    count = 0
    while 1:
        new_point = [antennae[0][0] + vdiff*count, antennae[0][1] + hdiff*count]
        if(is_valid(data, new_point)):
            points.append(new_point)
        else:
            break
        count += 1

    count = 0
    while 1:
        new_point = [antennae[1][0] - vdiff*count, antennae[1][1] - hdiff*count]

        if(is_valid(data, new_point)):
            points.append(new_point)
        else:
            break
        
        count += 1
    


    return(points)
```

```python
with open('data\\input.txt', 'r') as f:
    data = [list(i) for i in f.read().splitlines()]
    
with open('data\\test.txt', 'r') as f:
    test = [list(i) for i in f.read().splitlines()]
    
points = []
for row in range(len(data)):
    for col in range(len(data[row])):
        if(data[row][col] != '.'):
            for row_2 in range(row, len(data)):
                for col_2 in range(len(data[row_2])):
                    if(not (row == row_2 and col == col_2) and data[row_2][col_2] == data[row][col]):
                        new_points = [point for point in get_antinode_points(data, [[row, col], [row_2, col_2]]) if is_valid(data, point) and point not in points]
                        for point in new_points:
                            points.append(point)
len(points)
```
